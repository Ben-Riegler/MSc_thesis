```python
import torch
import stheno

from convcnp.architectures import SimpleConv, UNet
from convcnp.set_conv import ConvCNP
from convcnp.utils import device
import convcnp.data

from eval_fcts import plot_model_task
```

```python
%%capture
model_eq = ConvCNP(learn_length_scale=True,
                points_per_unit=64,
                architecture=UNet())

path_eq = "experiments/convcnpxl-eq/checkpoint.pth.tar"
load_dict = torch.load(path_eq)
model_eq.load_state_dict(load_dict['state_dict'])
model_eq.to(device)
```

```python
for i in range(16):


    n_ctxt = 1+i
    ctxt_ran = (-2,2)

    gen = convcnp.data.GPGenerator(x_range = ctxt_ran, kernel=stheno.EQ().stretch(0.25))
    task = gen.generate_task(n_ctxt = n_ctxt)

    idx = i
    
    if (i % 2 == 0): pred_ran = (-2,2) 
    
    else: pred_ran = (-4,4)

    plot_model_task(model_eq, task, idx, pred_ran, save = i, name = "eq")
```
